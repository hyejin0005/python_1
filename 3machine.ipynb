```python
import pandas as pd
df = pd.read_csv("c:/Users/H.J.LIM/Downloads/data/data/iris.csv")
```

```python
df.columns
```

```python
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
#X 행렬 y 는 백터
X = df[features]
y= df['class']
X
```

```python
#트레이닝/ 테스트 데이터 분할
#random 안돌리면 다른값 나옴
#x_tn-> train te test
from sklearn.model_selection import train_test_split
X_tn, X_te,y_tn,y_te = train_test_split(X,y,random_state =8)
```

```python
X_tn
```

#데이터 표준화 
한쪽이 큰게 있으면 한족으로 팍 올라가기 떄문에 standard해줌
standardscaler -> 평균이 0이고 분산이 1인 값으로 변경
=standadization(마이너스값도 나올수 있음)
normalization -> 0~1(마이너스 안나옴)/minmaxscale (0과 1사이)
regularization-> 제약조건 븥임
#fit(X_tn) -> 주어진 데이터 에서 평균과 표준편차 계산하는 역할
x-x(평균)/s  
#standardscaler.fit


```python
#데이터 표준화
from sklearn.preprocessing import StandardScaler
std_scale = StandardScaler()
std_scale.fit(X_tn)
X_tn_std = std_scale.transform(X_tn) #train에 평균빼고 표준편차로 나누면 -1,0,1 나옴
X_te_std = std_scale.transform(X_te)
```

```python
X_tn_std
```

```python
#학습
from sklearn.neighbors import KNeighborsClassifier
clf_knn = KNeighborsClassifier(n_neighbors=2)
clf_knn.fit(X_tn_std, y_tn)
#--------
#neighbors = k값임 
#clf_knn.fit(기출문제, 정답)
```

```python
#예측
knn_pred= clf_knn.predict(X_te_std)
print(knn_pred)
```
